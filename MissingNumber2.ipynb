```python
def missingNumber(nums):
    nums.sort()

    for number in range(len(nums)):
        if nums[number] != number:
            return number

    return len(nums)
```

```python
missingNumber(nums = [3,0,1])
```

```python
missingNumber(nums = [0,1])
```

```python
missingNumber(nums = [9,6,4,2,3,5,7,0,1])
```

```python
def missingNumber1(nums):
  hash_set = set(nums)

  for number in range(len(hash_set)):
    if number not in hash_set:
      return number
  return len(hash_set)
```

```python
missingNumber1(nums = [3,0,1])
```

```python
missingNumber1(nums = [0,1])
```

```python
missingNumber1(nums = [9,6,4,2,3,5,7,0,1])
```

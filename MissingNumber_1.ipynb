```python
from typing import List

def missingNumber(nums: List[int]) -> int:
        n = len(nums)

        for i in range(n + 1):
            if i not in nums:
                return i
```

```python
nums1 = [3,0,1]
missingNumber(nums1)
```

```python
nums2 = [0,1]
missingNumber(nums2)
```

```python
nums3 = [9,6,4,2,3,5,7,0,1]
missingNumber(nums3)
```

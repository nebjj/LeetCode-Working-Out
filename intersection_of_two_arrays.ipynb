```python
from typing import List

def intersect(nums1: List[int], nums2: List[int]) -> List[int]:
  inter_list = []
  dict_inter = {}
  for number in nums1:
    if number in dict_inter:
      dict_inter[number] += 1
    else:
      dict_inter[number] = 1

  for number in nums2:
    if number in dict_inter:
      dict_inter[number] -= 1
      if dict_inter[number] >= 0:
        inter_list.append(number)
  return inter_list
```

```python
intersect(nums1 = [1,2,3,4,4], nums2 = [4,4])
```

```python
intersect(nums1 = [1,2,3,4,4], nums2 = [3,4])
```
